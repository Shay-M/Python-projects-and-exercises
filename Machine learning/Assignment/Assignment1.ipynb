```python
# 4.a.
# ------------>>>>>>>> RUN THIS CODE CELL <<<<<<<<------------
# --------  (run after implementation)
# === CODE TYPE: ANSWER

def calc_distance(X_test, X_train):
    # YOUR CODE HERE
    X_test.index.name = 'test'
    X_train.index.name = 'train'
    distance_df = pd.DataFrame(0.0, columns=X_train.index ,index=X_test.index)
    
    for i_test, test_row in X_test.iterrows():
        for i_train, train_row in X_train.iterrows():
            distance_df.at[i_test, i_train] = ((test_row - train_row)**2).sum()**0.5
    return distance_df
    
```

```python
# 4.b.
# ------------>>>>>>>> RUN THIS CODE CELL <<<<<<<<------------
# --------  (run after implementation)
# === CODE TYPE: ANSWER 

def predict(X_test, X_train, y_train, k):
    # YOUR CODE HERE
    
    distance_df = calc_distance(X_test,X_train)
    pred = {}
    for i_dist,dist_row in distance_df.iterrows():
        val = y_train.loc[dist_row.nsmallest(k).index].value_counts().head(1).index[0]
        # nsmallest: to select the three rows having the smallest values in column " "
        # Series.value_counts:Return a Series containing counts of unique values.
        pred[i_dist] = val
    
    return pd.Series(pred,name='y_predicted') #y_test
```

```python
# 5.
# ------------>>>>>>>> RUN THIS CODE CELL <<<<<<<<------------
# --------  (run after implementation)
# === CODE TYPE: ANSWER 

def evaluate_performance(y_predicted ,y_test):
    # YOUR CODE HERE
    correct_pred = ((y_predicted == y_test)*1).sum()
    test_instances = y_predicted.count()
    accuracy = correct_pred/test_instances
    error_rate = 1 - accuracy
    return error_rate
```
