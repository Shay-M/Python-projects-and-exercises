```python
# נעשה חיתוך ונציג רק חלק של התמונה.
plt.imshow(np_sloar_grayscale.T[240:390,120:270],cmap='gray')
```

```python
# נעשה חיתוך ונציג רק חלק של התמונה.
plt.imshow(np_sloar_grayscale.T[240:390,120:270].T,cmap='gray')
```

```python
grayscale_mean,grayscale_min,grayscale_max = np.mean(np_sloar_grayscale), np.min(np_sloar_grayscale), np.max(np_sloar_grayscale)
grayscale_mean,grayscale_min,grayscale_max
```

```python
# שילוב של טכניקת mask עם np.where ייתן את התוצאה הבא:
# use where and masked np array
masked_photo = np.where(np_sloar_grayscale > grayscale_mean, 0, 1)
plt.imshow(masked_photo,cmap='gray')
```
